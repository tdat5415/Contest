```
import csv
import os
import pandas as pd
import pefile
import shutil
import sys
from glob import glob
from tqdm import tqdm
```

### **csv로 분류하는 함수**

```
def csv_classifier(base_path):
  with open(base_path + '/trainSet.csv', 'r', encoding='UTF-8-sig') as f:
    lines = f.readlines()
  data = list(csv.reader(lines))
  if data[0] == ['hash', 'result']: data = data[1:]
  data = sorted(data, key=lambda x: x[0])

  cdict = {
      '0x014c':[], '0x8664':[], '0xaa64':[], '0x01c4':[],
      '0x0166':[], '0x0184':[], 'NonPE' :[], 'elsePE':[]
  }
  for line in tqdm(data):
    try:
      pe = pefile.PE(base_path + '/trainSet/' + line[0])
      mach = pe.FILE_HEADER.Machine
      if    mach == 0x014c: cdict['0x014c'].append(line)
      elif  mach == 0x8664: cdict['0x8664'].append(line)
      elif  mach == 0xaa64: cdict['0xaa64'].append(line)
      elif  mach == 0x01c4: cdict['0x01c4'].append(line)
      elif  mach == 0x0166: cdict['0x0166'].append(line)
      elif  mach == 0x0184: cdict['0x0184'].append(line)
      else:                 cdict['elsePE'].append(line)
    except pefile.PEFormatError:
      cdict['NonPE'].append(line)

  for name in cdict.keys():
    with open(base_path + '/%s.csv'%name, 'w', newline='') as f:
      cw = csv.writer(f)
      for row in cdict[name]:
        cw.writerow(row)
  
  print(base_path + ' 분류완료')
```
